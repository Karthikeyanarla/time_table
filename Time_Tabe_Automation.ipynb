Write course id of your preferred course in the selection list to get the time table.

# **Output Instructions**
See the output at the end of the notebook (ofcourse after running all cells).
If two courses appear in the same cell of table
Then there is a clash.




```python
selection = ["CS 613", "HS 520", "EH 608", "MS 491", "CS 433"]
```

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)
url1 = "https://raw.githubusercontent.com/Karthikeyanarla/time_table/master/timeslots.csv"
url2 = "https://raw.githubusercontent.com/Karthikeyanarla/time_table/master/timetable.csv"
timing_dataframe = pd.read_csv(url1)
class_works = pd.read_csv(url2)
```

```python
dict_time = {}
list1 = ["Lecture", "Tutorial", "Lab"]
flag = 0
for i in selection:
  index = (class_works["Course Code"] == i).idxmax()
  row = class_works.iloc[index]
  course_name = row["Course Name"]
  for j in list1:
    node = row[j]
    if(type(node) != float):
      node_list = node.split("\n")
      if(len(node_list[0])):
        slot_name = node_list[0].split(",")
        for temp in slot_name:
          temp = temp[:2]
          if temp not in dict_time:
            dict_time[temp] = i + " " + course_name + "-" + j
          else:
            flag = 1
            dict_time[temp] +=  ", " + i + " " + course_name + "-" + j
if(flag):
  print("Choose wisely, there are some clashes")

dummy_dataframe = timing_dataframe.copy()
for rowIndex, row in timing_dataframe.iterrows(): 
    for columnIndex, value in row.items():
        if(value in dict_time):
          index = (timing_dataframe["Slot"] == row[0]).idxmax()
          dummy_dataframe[columnIndex][index] = dict_time[value]
dummy_dataframe
```
